```python
import numpy as np

original_list = [2,3,4,5,6,2,45,2,79]
original_list2 = [2,3,4,5,6,2,56,78,5]

def covariance(x,y):
    if len(x) != len(y):
        return "none"

    x_mean = np.mean(x)
    y_mean = np.mean(y)

    sum = 0

    for i in range(0, len(x)):
        sum += ((x[i] - x_mean) * (y[i] - y_mean))

    return sum/(len(x)-1)


print(covariance(original_list, original_list2))
np.cov(original_list, original_list)[0][1]
```

```python
import numpy as np

def cov_matrix_calculation(data):
    # calculate covariance matrix of the data
    cov_matx = np.cov(data.T)
    return cov_matx

dataset = np.array([[1, 1, 1], [1, 2, 1], [1, 3, 2], [1, 4, 3]])
print(cov_matrix_calculation(dataset))
```
